***<br>Removing outliers using 1.5 x IQR technique for all numerical columns<br>***

```python
cd_df.info()
```

```python
num_cols = cd_df.select_dtypes(include=['int', 'float']).columns.tolist()
```

```python
print(num_cols)
```

```python
Q_25 = cd_df[num_cols].quantile(0.25)
Q_75 = cd_df[num_cols].quantile(0.75)
IQR = Q_75 - Q_25
```

```python
Q_25
```

```python
Q_75
```

```python
IQR
```

```python
cd_df.head()
```

```python
Q_25['income']
```

```python
def remove_lower_outlier(dataFrame, col_name):
    threshold = Q_25[col_name]-(IQR[col_name]*1.5)
    return dataFrame[dataFrame[col_name] > threshold]
```

```python
def remove_higher_outlier(dataFrame, col_name):
    threshold = Q_75[col_name]+(IQR[col_name]*1.5)
    return dataFrame[dataFrame[col_name] < threshold]
```

```python
cd_df[cd_df.monthly_premium > (Q_75['monthly_premium']+(IQR['monthly_premium']*1.5))]
```

```python
cd_df = remove_higher_outlier(cd_df,'monthly_premium')
```

```python
cd_df = remove_lower_outlier(cd_df,'monthly_premium')
```

```python
cd_df = remove_higher_outlier(cd_df,'income')
cd_df = remove_higher_outlier(cd_df,'customer_lifetime_value')
cd_df = remove_higher_outlier(cd_df,'total_claim_amount')
cd_df = remove_lower_outlier(cd_df,'income')
cd_df = remove_lower_outlier(cd_df,'customer_lifetime_value')
cd_df = remove_lower_outlier(cd_df,'total_claim_amount')
```
