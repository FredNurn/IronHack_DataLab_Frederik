**<br>For the numerical variables, check for correlation between the input features.<br>** 

```python
numerical_df2 = numerical_df.drop(columns=['total_claim_amount'])
df3 = numerical_df2.corr()
df3.to_excel(r'data/Customer_Analysis_corrheatmap.xlsx', index=False)
```

**<br>Plot the Correlation Heatmap.<br>**

```python
plt.figure(figsize=(16, 6))
mask = np.triu(np.ones_like(numerical_df2.corr(), dtype='bool'))
heatmap = sns.heatmap(numerical_df2.corr(), mask=mask, vmin=-1, vmax=1, annot=True, cmap='BrBG')
heatmap.set_title('Triangle Correlation Heatmap', fontdict={'fontsize':18}, pad=16);
```

**<br>Show a plot of the response by the sales channel.<br>**

```python
index = list(range(0,len(basic_file)))
```

```python
basic_file['index']=index

response_channel = basic_file.groupby(['response','sales_channel'])['index'].count().reset_index()
response_channel

```

```python
plt.figure(figsize=(10,7))
plt.title('Response by Sales Channel', size='25', fontweight='bold')

sns.barplot(x = response_channel['sales_channel'],y = response_channel['index'], hue = response_channel['response'])
```
