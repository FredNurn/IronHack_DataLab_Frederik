**<br>For the numerical variables, check for correlation between the input features.<br>** 

```python
numerical_df2 = numerical_df.drop(columns=['total_claim_amount'])
numerical_df2.corr()
```

**<br>Plot the Correlation Heatmap.<br>**

```python
plt.figure(figsize=(16, 6))
mask = np.triu(np.ones_like(numerical_df2.corr(), dtype='bool'))
heatmap = sns.heatmap(numerical_df2.corr(), mask=mask, vmin=-1, vmax=1, annot=True, cmap='BrBG')
heatmap.set_title('Triangle Correlation Heatmap', fontdict={'fontsize':18}, pad=16);
```
