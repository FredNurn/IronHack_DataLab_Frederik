```python
ohe = OneHotEncoder(handle_unknown='ignore')
region_encoded = pd.DataFrame(ohe.fit_transform(categoricals_df[['region']]).toarray())
region_encoded.columns = ['central', 'east', 'north west', 'west']
```

```python
response_encoded = pd.DataFrame(ohe.fit_transform(categoricals_df[['response']]).toarray())
response_encoded.columns = ['no', 'yes']
```

```python
month_encoded = pd.DataFrame(ohe.fit_transform(categoricals_df[['month']]).toarray())
month_encoded.columns = ['feb', 'jan']
```

```python
gender_encoded = pd.DataFrame(ohe.fit_transform(categoricals_df[['gender']]).toarray())
gender_encoded.columns = ['m', 'f']
```

```python
marital_status_encoded = pd.DataFrame(ohe.fit_transform(categoricals_df[['marital_status']]).toarray())
marital_status_encoded.columns = ['divorced','married','single']
```

```python
policy_encoded = pd.DataFrame(ohe.fit_transform(categoricals_df[['policy']]).toarray())
policy_encoded.columns = ['corporate l1','corporate l2', 'corporate l3','personal l1','personal l2','personal l3','special l1','special l2','special l3']
```

```python
policy_type_encoded = pd.DataFrame(ohe.fit_transform(categoricals_df[['policy_type']]).toarray())
policy_type_encoded.columns = ['corporate auto', 'personal auto', 'special auto']
```

```python
renew_offer_type_encoded = pd.DataFrame(ohe.fit_transform(categoricals_df[['renew_offer_type']]).toarray())
renew_offer_type_encoded.columns = list(set(categoricals_df.renew_offer_type))
```

```python
sales_channel_encoded = pd.DataFrame(ohe.fit_transform(categoricals_df[['sales_channel']]).toarray())
sales_channel_encoded.columns = list(set(categoricals_df.sales_channel))
```

```python
vehicle_class_encoded = pd.DataFrame(ohe.fit_transform(categoricals_df[['vehicle_class']]).toarray())
vehicle_class_encoded.columns = list(set(categoricals_df.vehicle_class))
```

```python
enc = OrdinalEncoder(categories=[['high school or below','college', 'bachelor', 'master', 'doctor']])
categoricals_df2['education_coded'] = enc.fit_transform(categoricals_df2[['education']]).reshape((1,-1))[0]
```

```python
enc = OrdinalEncoder(categories=[[ 'basic','extended','premium']])
categoricals_df2['coverage_coded'] = enc.fit_transform(categoricals_df2[['coverage']]).reshape((1,-1))[0]
```

```python
enc = OrdinalEncoder(categories=[['disabled', 'unemployed','medical leave', 'retired', 'employed']])
categoricals_df2['employment_status_coded'] = enc.fit_transform(categoricals_df2[['employment_status']]).reshape((1,-1))[0]
```

```python
enc = OrdinalEncoder(categories=[[ 'rural','suburban','urban']])
categoricals_df2['location_code_coded'] = enc.fit_transform(categoricals_df2[['location_code']]).reshape((1,-1))[0]
```

```python
enc = OrdinalEncoder(categories=[[ 'small', 'medsize', 'large']])
categoricals_df2['vehicle_size'] = enc.fit_transform(categoricals_df2[['vehicle_size']]).reshape((1,-1))[0]
```
