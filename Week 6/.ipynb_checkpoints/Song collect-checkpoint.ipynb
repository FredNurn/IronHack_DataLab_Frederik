```python
import spotify_config
import spotipy
import json
from spotipy.oauth2 import SpotifyClientCredentials
import pprint
import pandas as pd
import openpyxl
```

```python
sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id= spotify_config.client_id, client_secret= spotify_config.client_secret))
```

```python
list_of_audio_features=[]
```

```python
def get_playlist_tracks(username, playlist_id):
    results = sp.user_playlist_tracks(username,playlist_id,market="DE")
    tracks = results['items']
    while results['next']:
        results = sp.next(results)
        tracks.extend(results['items'])
    return results
```

```python
playlists=["37i9dQZF1EQmPV0vrce2QZ","37i9dQZF1DWZBCPUIUs2iR","37i9dQZF1DX47i8ytbYGJE"]
```

```python
for playlist in playlists:
    playlist2 = sp.user_playlist_tracks("spotify", playlist)
    tracks = get_playlist_tracks('spotify',playlist)
    for i in range(len(playlist2['items'])):
         list_of_audio_features.append(sp.audio_features(tracks['items'][i]["track"]["id"])[0])
```

```python
df=pd.DataFrame(list_of_audio_features)    
df=df[["danceability","energy","loudness","speechiness","acousticness",
    "instrumentalness","liveness","valence","tempo","id","duration_ms"]]
```
